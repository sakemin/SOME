```python
from compressor import vocal_compressor
import numpy, librosa
```

```python
# Load audio file
file_path = "/home/sake/userdata/sake/a2m/hummings/guitar_humming_2.wav"  # Path to your vocal audio file
y, sr = librosa.load(file_path, sr=None)

import IPython.display as ipd

# Play original audio
print("Original audio:")
ipd.display(ipd.Audio(y, rate=sr))
```

```python
y_comp = vocal_compressor(
    y=y,
    sr=sr,
    threshold=-22.0,  # -22dB:   ,   
    ratio=2.5,       # 2.5:1:     
    attack=0.010,    # 10ms:  
    release=0.1,     # 100ms:     
    makeup_gain=2.0  # 2dB:   
)
```

```python
ipd.display(ipd.Audio(y_comp, rate=sr))
```
